# Home Credit Scorecard Model (Final Project)

#### Background Case

Homecredit currently is using any statisticals methods and machine learning to predicting the credit score. In this case, we can ensure customers who are able to make payments are not rejected when applying for a loan and loans can be provided with a principal, maturity, and repayment calendar that will motivate customers to succeed.

## Data Description

1. SK_ID_CURR &rarr; ID of loan in our sample
2. TARGET &rarr; Target variable (1 - client with payment difficulties: he/she had late payment more than X days on at least one of the first Y installments of the loan in our sample, 0 - all other cases)
3. NAME_CONTRACT_TYPE &rarr; Identification if loan is cash or revolving
4. CODE_GENDER &rarr; Gender of the client
5. FLAG_OWN_CAR &rarr; Flag if the client owns a car
6. FLAG_OWN_REALTY &rarr; Flag if client owns a house or flat
7. CNT_CHILDREN &rarr; Number of children the client has
8. AMT_INCOME_TOTAL &rarr; Income of the client
9. AMT_CREDIT &rarr; Credit amount of the loan
10. AMT_ANNUITY &rarr; Loan annuity
11. AMT_GOODS_PRICE &rarr; For consumer loans it is the price of the goods for which the loan is given
12. NAME_TYPE_SUITE &rarr; Who was accompanying client when he was applying for the loan
13. NAME_INCOME_TYPE &rarr; "Clients income type (businessman, working, maternity leave,â€¦)"
14. NAME_EDUCATION_TYPE &rarr; Level of highest education the client achieved
15. NAME_FAMILY_STATUS &rarr; Family status of the client
16. NAME_HOUSING_TYPE &rarr; "What is the housing situation of the client (renting, living with parents, ...)"
17. REGION_POPULATION_RELATIVE &rarr; Normalized population of region where client lives (higher number means the client lives in more populated region),normalized 
18. DAYS_BIRTH &rarr; Client's age in days at the time of application,time only relative to the application
19. DAYS_EMPLOYED &rarr; How many days before the application the person started current employment,time only relative to the application
20. DAYS_REGISTRATION &rarr; How many days before the application did client change his registration,time only relative to the application
21. DAYS_ID_PUBLISH &rarr; How many days before the application did client change the identity document with which he applied for the loan,time only relative to the application
22. OWN_CAR_AGE &rarr; Age of client's car
23. FLAG_MOBIL &rarr; "Did client provide mobile phone (1=YES, 0=NO)"
24. FLAG_EMP_PHONE &rarr; "Did client provide work phone (1=YES  0=NO)" 
25. FLAG_WORK_PHONE &rarr; "Did client provide home phone (1=YES  0=NO)" 
26. FLAG_CONT_MOBILE &rarr; "Was mobile phone reachable (1=YES  0=NO)" 
27. FLAG_PHONE &rarr; "Did client provide home phone (1=YES  0=NO)" 
28. FLAG_EMAIL &rarr; "Did client provide email (1=YES  0=NO)" 
29. OCCUPATION_TYPE &rarr; What kind of occupation does the client have 
30. CNT_FAM_MEMBERS &rarr; How many family members does client have 
31. REGION_RATING_CLIENT &rarr; "Our rating of the region where client lives (1 2 3)" 
32. REGION_RATING_CLIENT_W_CITY &rarr; "Our rating of the region where client lives with taking city into account (1 2 3)" 
33. WEEKDAY_APPR_PROCESS_START &rarr; On which day of the week did the client apply for the loan 
34. HOUR_APPR_PROCESS_START &rarr; Approximately at what hour did the client apply for the loan rounded
35. REG_REGION_NOT_LIVE_REGION &rarr; "Flag if client's permanent address does not match contact address (1=different  0=same  at region level)" 
36. REG_REGION_NOT_WORK_REGION &rarr; "Flag if client's permanent address does not match work address (1=different  0=same  at region level)" 
37. LIVE_REGION_NOT_WORK_REGION &rarr; "Flag if client's contact address does not match work address (1=different  0=same  at region level)" 
38. REG_CITY_NOT_LIVE_CITY &rarr; "Flag if client's permanent address does not match contact address (1=different  0=same  at city level)"

## Cleansing Data

### Create a pandas DataFrame  from object in Matplotlib

```python
# Show the data shape
print('Data shape :{}'.format(df.shape))
print('')
print('Data info :{}'.format(df.info()))
```

```python
print(df['CODE_GENDER'].value_counts())
```

```python
df = df[df['CODE_GENDER'] != 'XNA']
```

```python
gender_counts = df['CODE_GENDER'].value_counts()
print(gender_counts)
plt.bar(gender_counts.index, gender_counts.values)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution')
plt.show()
```

```python
contract_counts = df.groupby(['NAME_CONTRACT_TYPE', 'CODE_GENDER']).size().unstack()
print(contract_counts)
contract_counts.plot(kind='bar')
plt.xlabel('Contract Types')
plt.ylabel('Count')
plt.title('Payment Difficulties')
plt.show()
```

```python
target_counts = df.groupby(['TARGET', 'CODE_GENDER']).size().unstack()
print(target_counts)
target_counts.plot(kind='bar')
plt.xlabel('Target')
plt.ylabel('Count')
plt.title('Payment Difficulties')
plt.show()
```

```python
car_counts = df.groupby(['FLAG_OWN_CAR', 'CODE_GENDER']).size().unstack()
print(car_counts)
car_counts.plot(kind='bar')
plt.xlabel('Car Ownership')
plt.ylabel('Count')
plt.title('Car ownership by Gender')
plt.show()
```

```python
house_counts = df.groupby(['FLAG_OWN_REALTY', 'CODE_GENDER']).size().unstack()
print(house_counts)
house_counts.plot(kind='bar')
plt.xlabel('House Ownership')
plt.ylabel('Count')
plt.title('House ownership by Gender')
plt.show()
```
